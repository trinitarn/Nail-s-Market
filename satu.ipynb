```python
import pandas as pd
import matplotlib.pyplot as plt
!pip install scipy
import scipy
from scipy.stats import pearsonr
```

```python
df = pd.read_csv("income_spending.csv")

# Filter for male customers
male_data = df[df["Genre"] == "Male"]

# Calculate correlation coefficient
correlation_coef, _ = pearsonr(male_data['Age'], male_data['Spending Score (1-100)'])

# Print the correlation coefficient
print("Correlation coefficient:", correlation_coef)

# Create a scatter plot
plt.scatter(male_data['Age'], male_data['Spending Score (1-100)'])
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.title('Correlation Between Age and Spending Score for Male Customers')
plt.show()
```

```python
df = pd.read_csv("income_spending.csv")

# Filter for female customers
female_data = df[df["Genre"] == "Female"]

# Calculate correlation coefficient
correlation_coef, _ = pearsonr(female_data['Age'], female_data['Spending Score (1-100)'])

# Print the correlation coefficient
print("Correlation coefficient:", correlation_coef)

# Create a scatter plot
plt.scatter(female_data['Age'], female_data['Spending Score (1-100)'])
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.title('Correlation Between Age and Spending Score for Female Customers')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
!pip install scipy
import scipy
from scipy.stats import pearsonr
```

```python
df = pd.read_csv("cust_segment.csv")

# Filter for male customers
clothing_data = df[df["preferred_category"] == "Clothing"]

# Calculate correlation coefficient
correlation_coef, _ = pearsonr(clothing_data['purchase_frequency'], clothing_data['membership_years'])

# Print the correlation coefficient
print("Correlation coefficient:", correlation_coef)

# Create a scatter plot
plt.scatter(clothing_data['membership_years'], clothing_data['purchase_frequency'])
plt.ylabel('purchase_frequency')
plt.title('Correlation Between Membership Year and Purchase Frequency')
plt.show()
```

```python
import pandas as pd

# Assuming your data is in a CSV file named "campaign_data.csv"
data = pd.read_csv("Social_Media_Advertising.csv")

# Group the data by campaign goal and channel, then calculate the mean conversion rate
grouped_data = data.groupby(["Campaign_Goal", "Channel_Used"])["Conversion_Rate"].mean().reset_index()

# Find the best channel for each campaign goal
best_channels = grouped_data.groupby("Campaign_Goal").apply(lambda x: x.nlargest(1, "Conversion_Rate")).reset_index(drop=True)

# Print the results
print(best_channels)
```
